```python
import keras
import numpy as np
import shutil
import os
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
aug_data = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
```

```python
ignore = ['.DS_Store', 'space', 'del']
letters = os.listdir('datasets/asl-alphabet-test')
for letter in letters:
    if letter not in ignore:
        counter = 1
        os.mkdir('aug_dataset/' + letter)
        while counter < 30:
            if counter < 10:
                append = "0" + str(counter)
            else:
                append = str(counter)
            counter += 1
            current_dir = 'datasets/asl-alphabet-test/' + letter + '/' + letter + '00' + append + '_test.jpg'
            new_dir = 'aug_dataset/' + letter + '/' + letter + '00' + append + '_train.jpg'
            img = load_img(current_dir)
            shutil.copy(current_dir, new_dir)
            x = img_to_array(img)
            x = x.reshape((1,) + x.shape)
            i = 0
            try:
                os.mkdir('aug_images/' + letter)
            except:
                pass
            for batch in aug_data.flow(x, batch_size=1, 
                                      save_to_dir='aug_images/' + letter, save_prefix=letter, save_format='jpg'):
                i += 1
                if i > 17:
                    break
```

```python
append_aug_list = os.listdir('aug_images')
ignore = ['.DS_Store']
for letter in append_aug_list:
    if letter not in ignore:
        images_in_dir = os.listdir('aug_images/' + letter)
        counter = 30
        for image in images_in_dir:
            if counter < 100:
                append = letter + "00" + str(counter)
            else:
                append = letter + "0" + str(counter)
            counter += 1
            shutil.move('aug_images/' + letter + '/' + image, 'aug_dataset/' + letter + '/' + append + '_train.jpg')
```
